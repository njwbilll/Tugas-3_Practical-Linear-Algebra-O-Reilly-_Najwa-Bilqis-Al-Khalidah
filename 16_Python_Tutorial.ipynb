## Fungsi dan Control Flow

### Penjelasan Teori
Fungsi memungkinkan kita membungkus logika yang berulang, sementara *Control Flow* memungkinkan kode untuk mengambil keputusan atau melakukan iterasi. Dalam aljabar linear, kita harus menghindari *loops* (perulangan) sebanyak mungkin dan lebih mengandalkan *vectorization* dari NumPy karena jauh lebih cepat secara komputasi.



```python
# 2. Menulis Fungsi Sendiri
def jumlah_kuadrat(vektor):
    return np.sum(vektor**2)

print("Hasil fungsi custom:", jumlah_kuadrat(vektor))

# 3. Control Flow (If-Else)
nilai = 85
if nilai >= 80:
    status = "Lulus"
else:
    status = "Remedial"
print(f"Status: {status}")
```

## Mengukur Waktu Komputasi (Benchmarking)

### Penjelasan Teori
Dalam ilmu data, efisiensi kode sangatlah krusial. Menggunakan `time.time()` memungkinkan kita mengukur durasi eksekusi sebuah algoritma. Selalu bandingkan antara metode *looping* tradisional dengan metode *vectorization* NumPy; Anda akan menemukan bahwa metode vektor jauh lebih unggul karena optimasi di level bahasa tingkat rendah.

```python
data_besar = np.random.randn(1000000)

# Metode Loop
start = time.time()
total = 0
for x in data_besar:
    total += x**2
waktu_loop = time.time() - start

# Metode Vectorization
start = time.time()
total_vec = np.sum(data_besar**2)
waktu_vec = time.time() - start

print(f"Waktu Loop: {waktu_loop:.5f} detik")
print(f"Waktu Vektor: {waktu_vec:.5f} detik")
print(f"Vektor lebih cepat {waktu_loop/waktu_vec:.1f} kali lipat")
```
