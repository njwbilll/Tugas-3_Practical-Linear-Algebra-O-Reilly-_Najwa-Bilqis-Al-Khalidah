## Ringkasan Materi
SVD adalah dekomposisi matriks yang dapat diterapkan pada semua jenis matriks, tidak terbatas pada matriks persegi. SVD memecah matriks $A$ menjadi $U \Sigma V^T$. Komponen ini mengungkapkan struktur informasi data, memungkinkan reduksi dimensi, dan memberikan solusi yang sangat stabil secara numerik untuk masalah pseudoinverse.

### Topik yang Dibahas:
1. Konsep U, Sigma, dan V^T
2. Hubungan SVD dengan Eigenvalues
3. Representasi matriks sebagai susunan layer (Rank-1 Layers)
4. Moore-Penrose Pseudoinverse via SVD


```python
import numpy as np
from numpy import linalg as LA

# 1. SVD dalam Python
A = np.random.randn(5, 3)
U, S, VT = LA.svd(A, full_matrices=False)

print("Bentuk U:", U.shape)
print("Nilai Singular (S):", S)
print("Bentuk VT:", VT.shape)
```

***
## Rank-1 Layers dan Kompresi

### Penjelasan Teori
Matriks $A$ dapat dipandang sebagai penjumlahan dari beberapa matriks berperingkat satu (rank-1 layers). Jika kita hanya mengambil beberapa komponen singular terbesar, kita melakukan kompresi data dengan tetap mempertahankan esensi informasi matriks tersebut.



```python
# Konstruksi matriks A dari layer pertama
layer1 = S[0] * np.outer(U[:, 0], VT[0, :])

print("Layer pertama (Rank-1) dari matriks A:\n", layer1)
```

***
## Pseudoinverse via SVD

### Penjelasan Teori
Pseudoinverse Moore-Penrose dapat dihitung secara langsung dari komponen SVD. Jika $A = U \Sigma V^T$, maka $A^+ = V \Sigma^+ U^T$, di mana $\Sigma^+$ adalah kebalikan dari nilai singular yang tidak nol. Ini memberikan solusi yang jauh lebih stabil dibandingkan metode invers matriks standar untuk matriks yang mendekati singular atau tidak persegi.

```python
# Menghitung Pseudoinverse menggunakan SVD
S_inv = np.diag(1/S)
A_pinv = VT.T @ S_inv @ U.T

print("Pseudoinverse dihitung via SVD:\n", A_pinv)
print("\nVerifikasi dengan np.linalg.pinv:\n", LA.pinv(A))
```
